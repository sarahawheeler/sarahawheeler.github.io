```python
import altair as alt
import pandas as pd
import tables
```

```python
data = pd.read_csv(r'building_inventory.csv')
data.head(5)
```

```python
hometown_data = data[data["County"] == "Fulton"]
hometown_data.head(5)
hometown_data["Agency Name"].value_counts()
```

```python
hometown_data["City"].unique()
```

```python
input_dropdown = alt.binding_select(options=[None, 'Astoria', 'Lewistown', 'Canton', 'Glasford'], labels=['All', 'Astoria', 'Lewistown', 'Canton', 'Glasford'],
                                    name='City')
selection = alt.selection_point(fields=['City'], bind=input_dropdown)

chart = alt.Chart(hometown_data).mark_bar().encode(
    alt.X("Agency Name:N"),
    y='count()',
    color = alt.Color("count():Q", scale=alt.Scale(scheme='blues'))
).properties(width = 700, height = 700).interactive().add_params(
    selection
).transform_filter(
    selection
)

chart
```

```python
chart.properties(width='container').save("hw6_chart1.json")
```

```python
hometown_data['Year Acquired'].describe()
```

```python
filtered_hw_data = hometown_data[(hometown_data['Year Acquired'] != 0) & (hometown_data['Year Constructed'] != 0)]
```

```python
filtered_hw_data["Year Acquired"].describe()
filtered_hw_data["Year Constructed"] = pd.to_datetime(filtered_hw_data["Year Constructed"], format='%Y', errors='coerce')
filtered_hw_data["Year Acquired"] = pd.to_datetime(filtered_hw_data["Year Acquired"], format='%Y', errors='coerce')
```

```python
input_dropdown2 = alt.binding_select(options=[None, 1, 2, 3, 4, 5], labels=['All', '1', '2', '3', '4', '5'],
                                    name='Total Floors')
selection2 = alt.selection_point(fields=['Total Floors'], bind=input_dropdown2)

chart2 = alt.Chart(filtered_hw_data).mark_point().encode(
    alt.X('Year Constructed:T', scale=alt.Scale(domain=['1830-01-01', '2015-01-01'])),
    alt.Y('Year Acquired:T', scale = alt.Scale(domain = ['1940-01-01', '2015-01-01'])),
    color = alt.Color('Total Floors:O', scale=alt.Scale(scheme='viridis'))
).properties(width = 700, height = 500).interactive().add_params(
    selection2
).transform_filter(selection2)

chart2
```

```python
chart2.properties(width='container').save("hw6_chart2.json")
```
